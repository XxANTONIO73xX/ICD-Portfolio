## Ingeniería de características

En este apartado realizamos varios procesos de extracción de características con el objetivo de facilitar a nuestros algoritmos de inferencias el análisis de estos.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in ["sex","hear_right","hear_left", "DRK_YN"]:
    df.loc[:, [col]] = le.fit_transform(df[[col]])
df[["sex","hear_right","hear_left", "DRK_YN"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal = OrdinalEncoder()
for col in ["urine_protein","SMK_stat_type_cd"]:
    df.loc[:, [col]] = ordinal.fit_transform(df[[col]])
df[["urine_protein","SMK_stat_type_cd"]]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
for col in ["age","height"]:
    df.loc[:, [col]] = scaler.fit_transform(df[[col]])
df[["age","height"]]
```

Se realiza un MinMaxScaler en las features age y height bajo la hipótesis de que al estar redondeadas de 5 en 5. Podemos existir valores entre estos. (e.g. entre 5 y 10 existen los valores 6, 7, 8, 9).

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
columns_to_encode = ["urine_protein", "SMK_stat_type_cd"]
df[columns_to_encode] = encoder.fit_transform(df[columns_to_encode])
```

```python
from sklearn.preprocessing import RobustScaler
robust = RobustScaler()
columnas = ['weight', 'waistline', 'sight_left', 'sight_right','SBP', 'DBP', 'BLDS', 'tot_chole', 'HDL_chole', 'LDL_chole', 'triglyceride', 'hemoglobin','serum_creatinine', 'SGOT_AST', 'SGOT_ALT', 'gamma_GTP']
for col in columnas:
    df.loc[:, [col]] = robust.fit_transform(df[[col]])
df[columnas]
```

Para las variables numéricas continuas, se utiliza el método RobustScaler. Esto se debe a que RoubustScaler si toma en cuanto los outliers en el reescalado de datos. Por lo que con esto podemos incluir estos outliers en nuestros conjuntos de entrenamiento y validación.

## Modelado de Clasificadores

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, recall_score, confusion_matrix
```

### Recortando datos

```python
df_Y = df.loc[df["DRK_YN"]==1]
df_Y = df_Y[:3000]
df_N = df.loc[df["DRK_YN"]==0]
df_N = df_N[:3000]
```

**Debido al costo computación que representa el análisis de las más de 900 mil instancias, realizaremos un recorte a 3000.**

```python
dataset = pd.concat([df_Y, df_N],axis=0)
dataset = dataset.sample(frac=1).reset_index(drop=True)
dataset
```

```python
Y = dataset["DRK_YN"]  # Variable objetivo
X = dataset.drop(columns=["DRK_YN"])  # Características
```

```python
Y.dtype
```

### Utilizando todas las *features*

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

names = [
    "Nearest Neighbors",
    "Linear SVM",
    "RBF SVM",
    "Gaussian Process",
    "Decision Tree",
    "Random Forest",
    "Neural Net",
    "AdaBoost",
    "Naive Bayes",
    "QDA",
]

modelos = [
    KNeighborsClassifier(n_neighbors=3),
    SVC(kernel='linear'),
    SVC(gamma=2, C=1),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
    MLPClassifier(alpha=1, max_iter=1000),
    AdaBoostClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis()
]
metricas_df = pd.DataFrame(columns=['Modelo', 'Exactitud', 'Sensibilidad', 'Especificidad'])
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import KFold
import warnings

warnings.filterwarnings("ignore", category=UserWarning)

# Número de divisiones en la validación cruzada
num_folds = 10

# Inicializa un objeto KFold
kf = KFold(n_splits=num_folds)

# Itera a través de tus modelos
for nombre, modelo in zip(names, modelos):
    scores = cross_val_score(modelo, X, Y, cv=kf, scoring='accuracy')
    sensibilidades = []
    especificidades = []

    # Realiza la validación cruzada y obtiene las predicciones
    y_pred = cross_val_predict(modelo, X.to_numpy(), Y.to_numpy(), cv=kf)
    
    # Calcular la matriz de confusión
    conf_matrix = confusion_matrix(Y, y_pred)
    
    # Calcular sensibilidad y especificidad
    sensibilidad = conf_matrix[1, 1] / (conf_matrix[1, 1] + conf_matrix[1, 0])
    especificidad = conf_matrix[0, 0] / (conf_matrix[0, 0] + conf_matrix[0, 1])

    # Agregar métricas al DataFrame
    metrics = {
        'Modelo': nombre,
        'Exactitud': scores.mean() * 100,
        'Sensibilidad': sensibilidad * 100,
        'Especificidad': especificidad * 100
    }
    metricas_df = pd.concat([metricas_df, pd.DataFrame(metrics, index=[0])], ignore_index=True)

    # Crear un heatmap de la matriz de confusión
    print("Modelo:"+nombre)
    print("Matriz de confusión")
    print(conf_matrix)

metricas_df
```

### Aplicando Reducción de Dimensionalidad (PCA)

```python
from sklearn.decomposition import PCA

# Número de divisiones en la validación cruzada
num_folds = 10
pca = PCA(n_components=2)

# Aplica PCA a tus datos
X_pca = pca.fit_transform(X)

# Inicializa un objeto KFold
kf = KFold(n_splits=num_folds)

# Itera a través de tus modelos
for nombre, modelo in zip(names, modelos):
    scores = cross_val_score(modelo, X_pca, Y, cv=kf, scoring='accuracy')
    sensibilidades = []
    especificidades = []

    # Realiza la validación cruzada y obtiene las predicciones
    y_pred = cross_val_predict(modelo, X_pca, Y.to_numpy(), cv=kf)
    
    # Calcular la matriz de confusión
    conf_matrix = confusion_matrix(Y, y_pred)
    
    # Calcular sensibilidad y especificidad
    sensibilidad = conf_matrix[1, 1] / (conf_matrix[1, 1] + conf_matrix[1, 0])
    especificidad = conf_matrix[0, 0] / (conf_matrix[0, 0] + conf_matrix[0, 1])

    # Agregar métricas al DataFrame
    metrics = {
        'Modelo': nombre,
        'Exactitud': scores.mean() * 100,
        'Sensibilidad': sensibilidad * 100,
        'Especificidad': especificidad * 100
    }
    metricas_df = pd.concat([metricas_df, pd.DataFrame(metrics, index=[0])], ignore_index=True)

    # Crear un heatmap de la matriz de confusión
    print("Modelo:"+nombre)
    print("Matriz de confusión")
    print(conf_matrix)

metricas_df
```

Aplicando reducción de dimensionalidad (apartir de la instancia 10 del dataframe), podemos observar que los resultados son más bajos que sin aplicar este ultimo. En la siguente tabla se muestra una compración.
